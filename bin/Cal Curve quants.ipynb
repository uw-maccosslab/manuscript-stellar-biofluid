```python
# For calculating limit of quant
import matplotlib.pyplot as plt
from matplotlib import colors
import matplotlib as mpl
import numpy as np
import pandas as pd
import sys
import re
import os
from enum import Enum
#sys.path.append('C:\\Users\\philip.remes\\OneDrive - Thermo Fisher Scientific\\Documents\\Software\\Jupyter\\PythonUtilities\\')
import utilities_pmr as pmr
my_colors = pmr.setup_environment()
import calculate_loq as calc_loq
from collections import namedtuple
Metric = namedtuple('Metric', ['input_file', 'file_conc_map', 'label', 'output_file'])
files = [
    Metric('OT_DIA_MMCC_transition_areas_fixedna.csv',
           'OT_DIA_MMCC_file_replicates.csv',
           'OT_DIA', 'OTDIA_loq_boot_metrics'),
]

for info in files:
    print('Calculating for',info.input_file)
    config = calc_loq.get_default_config()
    # get the information for this experiment
    print('Loading Data')
    data_for_each_peptide = calc_loq.load_and_package_data(info.input_file, info.file_conc_map, normalize_to_global_stds=True)
    
    # calculate the loqs without optimizing the transitions
    print('Calculating Non-Optimized LOQs')
    nonopt_loq_info = calc_loq.calculate_quant_limit_for_peptides(data_for_each_peptide, config)
    # write the information to a file so we don't have to do the calculations again
    calc_loq.write_quant_limit_information(info.output_file + '_nonopt.txt', data_for_each_peptide, nonopt_loq_info)
    print('Calculating Optimized Transition LOQs')
    opt_loq_info = calc_loq.optimize_transitions_for_peptides(data_for_each_peptide, config)
    calc_loq.write_quant_limit_information(info.output_file + '_opt.txt', data_for_each_peptide, opt_loq_info)
    # write the new transition list
    calc_loq.save_opt_transition_information_for_skyline(info.output_file + '_opt_trans.txt', data_for_each_peptide, opt_loq_info)
```
